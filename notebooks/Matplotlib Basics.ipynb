### FIGURES AND SUBPLOTS

```python
#Â plots in matplotlib reside within a Figure obj
fig = plt.figure()
```

```python
# figsize - fix size and aspect ratio of the figures to be saved on disk
```

```python
# figure should be 2 X 2, selecting the 1st of 4 subplots
fig = plt.figure()
ax1 = fig.add_subplot(2, 2, 1)
ax2 = fig.add_subplot(2, 2, 2)
ax3 = fig.add_subplot(2, 2, 3)
```

```python
fig = plt.figure()
ax1 = fig.add_subplot(2, 2, 1)
ax2 = fig.add_subplot(2, 2, 2)
ax3 = fig.add_subplot(2, 2, 3)

plt.plot([1.5, 3.5, -2, 1.6])
```

```python
# 'k--' - style option - plots a black dashed line
fig = plt.figure()
ax1 = fig.add_subplot(2, 2, 1)
ax2 = fig.add_subplot(2, 2, 2)
ax3 = fig.add_subplot(2, 2, 3)

plt.plot(np.random.randn(50).cumsum(), 'k--')
```

```python
#Â objs returned by fig.add_subplot are AxesSupport objs
# one can directly plot on the other empty subplots by calling each one's instance method
fig = plt.figure()
ax1 = fig.add_subplot(2, 2, 1)
ax2 = fig.add_subplot(2, 2, 2)
ax3 = fig.add_subplot(2, 2, 3)

_ = ax1.hist(np.random.randn(100), bins=20, color='k', alpha=0.3)

ax2.scatter(np.arange(30), np.arange(30) + 3 * np.random.randn(30))

plt.plot(np.random.randn(50).cumsum(), 'k--')
```

```python
# plt.subplots - creates a figure and returns a NumPy array containing the created subplot objects
fig, axes = plt.subplots(2, 3)
axes
```

#### ADJUSTING THE SPACING AROUND SUBPLOTS

```python
# subplots_adjust on Figure obj - adjust the spacing b/w subplots
```

```python
# example - shrink the spacing to zero
fig, axes = plt.subplots(2, 2, sharex=True, sharey=True)
for i in range(2):
    for j in range(2):
        axes[i, j].hist(np.random.randn(500), bins=50, 
                        color='k', alpha=0.5)
plt.subplots_adjust(wspace=0, hspace=0)
```

### COLORS, MARKERS, AND LINE STYLES

```python
# plot x v/s y with green dashes
fig = plt.figure()
x = range(10)
y = range(10)
plt.plot(x, y, 'g--')
```

```python
#Â not munging strings for desired style
fig = plt.figure()
x = range(10)
y = range(10)

plt.plot(x, y, linestyle='--', color='r')
```

```python
# using hexcode colors
# CECECE - hexcode for light gray
fig = plt.figure()
x = range(10)
y = range(10)

plt.plot(x, y, linestyle='--', color='#CECECE')
```

```python
# line plots can have markets to highlight the actual points of data
from numpy.random import randn
```

```python
plt.plot(randn(30).cumsum(), 'ko--')
```

```python
# more explicitly, this could be written as
plt.plot(randn(30).cumsum(), color='k', linestyle='dashed', marker='o')
```

```python
data = np.random.randn(30).cumsum()
```

```python
plt.plot(data, 'k--', label='Default')
```

```python
plt.plot(data, 'k-', drawstyle='steps-post', label='steps-post')
```

```python
# subsequent points are linearly interpolated by default
data = np.random.randn(30).cumsum()

plt.plot(data, 'k--', label='Default')

plt.plot(data, 'k-', drawstyle='steps-post', label='steps-post')

plt.legend(loc='best')
```

```python
# calling plt.legend is necessary whether or not the label option is passed when plotting the data
```
