#### ADJUSTING THE SPACING AROUND SUBPLOTS

```python
# subplots_adjust on Figure obj - adjust the spacing b/w subplots
```

```python
# example - shrink the spacing to zero
fig, axes = plt.subplots(2, 2, sharex=True, sharey=True)
for i in range(2):
    for j in range(2):
        axes[i, j].hist(np.random.randn(500), bins=50, 
                        color='k', alpha=0.5)
plt.subplots_adjust(wspace=0, hspace=0)
```

#### SETTING THE TITLE, AXIS LABELS, TICKS, AND TICKLABELS

```python
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)

ax.plot(np.random.randn(1000).cumsum())
```

```python
#Â change x-axis ticks by using set_xticks and set_xlabels
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)

ax.plot(np.random.randn(1000).cumsum())

ticks = ax.set_xticks([0, 250, 500, 750, 1000])

labels = ax.set_xticklabels(['zero', 'two-fifty', 'five-hunds', 
                        'seven-fifty', 'thousand'], rotation=40,
                       fontsize='small')

#Â set title for the figure
ax.set_title('First Great Plot')

# set label
ax.set_label('Stages')
```

```python
# set - allows batch setting of plot props
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)

ax.plot(np.random.randn(1000).cumsum())

ticks = ax.set_xticks([0, 250, 500, 750, 1000])

labels = ax.set_xticklabels(['zero', 'two-fifty', 'five-hunds', 
                        'seven-fifty', 'thousand'], rotation=40,
                       fontsize='small')

props = {
    'title': 'First Great Plot Indeed!',
    'xlabel': 'Stages'
}

ax.set(**props)
```

### ANNOTATIONS AND DRAWING ON A SUBPLOT

```python
# add annotations using text, arrow, and annotate functions
# text - draws text at a given coordinates(x, y) on the plot
```

```python
#Â S&P 500 index price since 2007
#Â annotate important events
import pandas as pd
import numpy as np
from datetime import datetime
```

```python
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)

data = pd.read_csv('../resources/examples/spx.csv', index_col=0,
                   parse_dates=True)
spx = data['SPX']

spx.plot(ax=ax, style='k-')

crisis_data = [
    (datetime(2007, 10, 11), 'Peak of bull market'),
    (datetime(2008, 3, 12), 'Bear Stearns Fails'),
    (datetime(2008, 9, 15), 'Lehman Bankruptcy')
]

for date, label in crisis_data:
    ax.annotate(label, xy=(date, spx.asof(date) + 75),
                xytext=(date, spx.asof(date) + 225),
                arrowprops=dict(facecolor='black', headwidth=4,
                                width=2, headlength=4), 
                                horizontalalignment='left',
                                verticalalignment='top')

# zoom in on 2007-2010
ax.set_xlim(['1/1/2007', '1/1/2011'])
ax.set_ylim([600, 1800])

# set label
ax.set_title('Important dates in the 2008-2009 financial crisis')
```

```python
#Â ax.annonate - draw labels at indicated x and y co-ordinates
# set_xlim & set_ylim - set methods to start and end boundaries for plot
#Â ax.set_title - add title to the figure
```

```python
# patches - objects that represent common shapes
#Â ex. Rectangle and Circle
```

```python
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)

rect = plt.Rectangle((0.2, 0.75), 0.4, 0.15, color='k', alpha=0.3)
circ = plt.Circle((0.7, 0.2), 0.15, color='b', alpha=0.3)
pgon = plt.Polygon([[0.15, 0.15], [0.35, 0.4], [0.2, 0.6]],
                    color='g', alpha=0.5)

ax.add_patch(rect)
ax.add_patch(circ)
ax.add_patch(pgon)
```
